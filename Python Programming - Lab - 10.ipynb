### 08) WAP to create an UDF named frequency to count occurances of the specific word in a given text file.

```python
def frequency(file,word):
    with open(file,"r") as df:
        ans = df.read().lower().split().count(word)
        print(ans)
        
word = input("Enter any word :")
frequency("demo.txt",word)
```

### 10) WAP to write first 100 prime numbers to a file named primenumbers.txt 
(Note: each number should be in new line)5

```python
counter = 0
temp = 2
l = []

while counter != 100:
    for i in range(2,temp):
        if(temp % i == 0):
            break
    else:
        l.append(temp)
        counter += 1
    temp += 1

with open("primenumbers.txt", "w") as df:
    df.write(str(l))
```

### 12) WAP to replace word1 by word2 of a text file. Write the updated data to new file.

```python
firstWord = input("Enter first word :")
secondWord = input("Enter second word :")

with open("demo.txt", "r+") as df:
    ans = df.read()
    newans = ans.replace(firstWord, secondWord)
    df.write(newans)
```
