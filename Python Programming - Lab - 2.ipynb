### 12) WAP to accept date in dd mm yyyy format and display in multiple formats.
#### Input : 01 12 2025
#### Output :
##### 01/12/2025
##### 2025-12-01

```python
date = input("Enter date :")
date1 = date.replace(" ","/")
first,second,third = date.split(' ')
date2 = third+"-"+second+"-"+first

print(date1)
print(date2)
```

### 14) WAP to convert degree into Fahrenheit and vice versa.

```python
degree = float(input("Enter temperature in degree :"))
print(f"Your answer is {(degree * 9/5) + 32}f")

fahrenheit = float(input("Enter temperature in fahrenheit :"))
print(f"Your answer is {(fahrenheit - 32) * 5 / 9}d")
```
