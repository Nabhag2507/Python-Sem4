### 10) WAP to print GCD of given two numbers.

```python
a = int(input("Enter first number :"))
b = int(input("Enter second number :"))

ans = 0
min = a if a < b else b

for i in range(min+1,1,-1):
    if(a % i == 0 and b % i == 0):
        ans = i
        break
else:
    ans = 1
print(ans)
```

### 12) WAP to find whether the given number is Prime or not.

```python
n = int(input("Enter any number :"))
counter = 0
for i in range(2,n):
    if(n % i == 0):
        counter += 1
if(counter == 0):
    print("Number is prime")
else:
    print("Number is not prime")
```

### 13) WAP to print sum of digits of given number.

```python
n = int(input("Enter any number :"))

ans = 0
while(n > 0):
    ans += (n % 10)
    n = int(n / 10)
print(ans)
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
n = int(input("Enter any number :"))
ans = n
newN = 0
while(n > 0):
    temp = n % 10
    newN = (newN * 10) + temp
    n = int(n / 10)

if(ans == newN):
    print("Number is palindrome")
else:
    print("Number is not palindrome")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
num = int(input("Enter any number :"))
ans = num
newN = 0

while num > 0:
    temp = num % 10
    newN += (temp ** len(str(ans)))
    num = int(num / 10)

if ans == newN:
    print("Number is Armstrong")
else:
    print("Number is not Armstrong")
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
n = int(input("Enter any number :"))
ans = n
li = []

for i in range(1,n+1):
    temp = 0
    for j in range(1,i):
        if(i % j == 0):
            temp += j
    if(temp == i):
        li.append(i)
print(li)
```
