### 13) WAP to find second largest number in a list.

```python
l = list(map(int, input("Enter a list").split()))
l.sort(reverse=True)

for i in range(len(l)):
    if(l[i] != l[i+1]):
        print(l[i+1])
        break
else:
    print(l[0])
```

### 14) WAP to extract elements with frequency greater than K.

```python
l = list(map(int, input("Enter a list").split()))
k = int(input("Enter any number :"))

counter = 0
for i in set(l):
    if l.count(i) >= k:
        print(i,end=" ")
```
