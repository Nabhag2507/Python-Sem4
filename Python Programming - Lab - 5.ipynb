### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
s = input("Enter any string :")
uppercase = lowercase = digits = special = 0

for i in s:
    if(i.isupper()):
        uppercase += 1
    elif(i.islower()):
        lowercase += 1
    elif (i.isdigit()):
        digits += 1
    else:
        special += 1
print(f"Uppercase : {uppercase}, Lowercase : {lowercase}, Digits : {digits}, Special characters : {special}")
```

### 4) First and Last Occurrence Finder
- `programming`, `g` → First:3 Last:10
- `banana`, `a` → First:1 Last:5
- `hello`, `z` → Not found

```python
s = input("Enter any string :")
c = input("Enter any character :")

for i in range(len(s)):
    if(s[i] == c):
        print(f"First : {i}")
        break
for i in range(len(s)):
    if(s[len(s) - i - 1] == c):
        print(f"last : {len(s) - i - 1}")
        break
else:
    print("Not found")
```

### 8) Replace Vowels
- `Programming` → `Pr*gr*mm*ng`
- `Education` → `*d*c*t**n`
- `Sky` → `Sky`

```python
s = input("Enter any string")
ans = ""

for i in s:
    if(i.lower() in ['a', 'e', 'i', 'o', 'u']):
        ans += '*'
    else:
        ans += i

print(ans)
```

### 9) String Compression
### consecutive
- `aaabbcccc` → `a3b2c4`
- `xxxyyyz` → `x3y3z1`
- `abcd` → `a1b1c1d1`

```python
s = input("Enter any string :")
ans = ""
count = 1
for i in range(1,len(s)):
    if s[i-1] == s[i] :
        count += 1

    else:
        ans = ans + s[i-1] + str(count)
        count = 1
ans = ans + s[i-1] + str(count)
print(ans)
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
s = input("Enter any string")
ans = ""

for i in s:
    if(i.isupper()):
        ans += i.lower()
    else:
        ans += i.upper()

print(ans)
```
