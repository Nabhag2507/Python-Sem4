### 10) WAP to test if the given tuple is distinct or not.
**Sample Input:**
```
tuple1 = (1, 2, 3, 4)
tuple2 = (1, 2, 2, 3)
```
**Sample Output:**
```
(1, 2, 3, 4) is distinct? -> True
(1, 2, 2, 3) is distinct? -> False
```

````python
t = tuple(map(int,input("Enter tuple items :").split()))

t = sorted(t)
for i in range(len(t) - 1):
    if t[i] == t[i + 1]:
        print("List is not distinct")
        break
else:
    print("List is distinct")
````
