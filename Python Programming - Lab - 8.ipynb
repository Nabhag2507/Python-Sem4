### 01) WAP to find their union, intersection, and difference of the given two sets.

### 05) WAP to Merge two dictionaries. If a key exists in both dictionaries, add their values.

**Sample Input:**
```
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

```
**Sample Output:**
```
{1: 10, 2: 60, 3: 80, 4: 60}
```

````python
n = int(input("Enter size of dictonaries :"))
d1 = {}
print("Enter element for first dictonary")
for i in range(n):
    key = int(input("Enter key :"))
    value = int(input("Ente value :"))
    d1[key] = value

d2 = {}
print("Enter element for second dictonary")
for i in range(n):
    key = int(input("Enter key :"))
    value = int(input("Ente value :"))
    d2[key] = value

for i in d1.keys():
    if i not in d2.keys():
        d2[i] = d1[i]
    else:
        d2[i] = d1[i] + d2[i]
        
d2
````

### 06) WAP to Sort a dictionary in ascending order of its values.
Hint: in sorted(), use udf for key parameter


**Sample Input:**
```
{'Math': 70, 'Physics': 85, 'Chemistry': 60}

```
**Sample Output:**
```
{'Chemistry': 60, 'Math': 70, 'Physics': 85}
```

````python
def getvalue(item):
    return item[1]

n = int(input("Enter size of dictionary: "))
d = {}

for i in range(n):
    key = str(input("Enter subject name: "))
    value = int(input("Enter marks: "))
    d[key] = value

sorted_items = sorted(d.items(), key=getvalue)
print(dict(sorted_items))
````

### 07) WAP to find the keys having maximum and minimum values in a dictionary.

**Sample Input:**
```
{'A': 50, 'B': 90, 'C': 30}
```
**Sample Output:**
```
Max Value Key: B
Min Value Key: C

```

````python
data_dict = {'A': 50, 'B': 90, 'C': 30}

def find_max_min_keys(dictionary):
    max_key = max(dictionary, key=dictionary.get)
    min_key = min(dictionary, key=dictionary.get)
    return max_key, min_key
max_key, min_key = find_max_min_keys(data_dict)

print(f"Max Value Key: {max_key}")
print(f"Min Value Key: {min_key}")
````

### 09) WAP to count the frequency of each character in a string using a dictionary.
**Sample Input:**
```
programming
```
**Sample Output:**
```
{'p': 1, 'r': 2, 'o': 1, 'g': 2, 'a': 1, 'm': 2, 'i': 1, 'n': 1}
```

### 10) WAP to swap keys and values of a dictionary.
**Sample Input:**
```
{1: 'A', 2: 'B', 3: 'C'}
```
**Sample Output:**
```
{'A': 1, 'B': 2, 'C': 3}
```

````python
n = int(input("Enter size of dictonary :"))
d = {}

for i in range(n):
    key = int(input("Enter subject name :"))
    value = int(input("Enter marks of subject :"))
    d[key] = value

reverseDict = {value: key for key, value in d.items()}
reverseDict
````
