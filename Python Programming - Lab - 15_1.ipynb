### 03) WAP to create Circle class with area and perimeter function to find area and perimeter of circle.

### 05) Create a bank account class with methods to deposit, withdraw, and check balance.

```python
class bank:
    def __init__(self, balance = 0):
        self.balance = balance
        
    def deposit(self, amount):
        self.balance += amount
    
    def withdraw(self, amount):
        if self.balance <= amount:
            return "Insufficient balance"
        self.balance -= amount
        
    def check(self):
        return self.balance
    
user = bank()

user.deposit(5000)
user.withdraw(1500)
user.check()
```

### 06) Create a class for managing inventory that includes attributes such as item name, price, quantity, and methods to add, remove, and update items.

```python
class Inventory:
    def __init__(self):
        self.inventory = {} 
        
    def add_item(self, name, price, quantity):
        if name in self.inventory:
            self.inventory[name]['quantity'] += quantity
        else:
            self.inventory[name] = {'price': price, 'quantity': quantity}

    def remove_item(self, name, quantity):
        if name in self.inventory:
            self.inventory[name]['quantity'] -= quantity
            if self.inventory[name]['quantity'] <= 0:
                del self.inventory[name]
        else:
            print("Item not found.")

    def update_item(self, name, price=None, quantity=None):
        if name in self.inventory:
            if price is not None: self.inventory[name]['price'] = price
            if quantity is not None: self.inventory[name]['quantity'] = quantity
        else:
            print("Item not found.")

    def display(self):
        for name, details in self.inventory.items():
            print(f"{name}: {details}")

inv = Inventory()
inv.add_item("Apple", 0.5, 100)
inv.display()

inv.update_item("Apple", price=0.6)
inv.remove_item("Apple", 10)
inv.display()
```

### 09) Define Time class with hour and minute as data member. Also define addition method to add two time objects.

```python
class time:
    def __init__(self, h, m):
        self.hour = h
        self.minute = m

    @classmethod
    def addition(cls, t1, t2):
        h = t1.hour + t2.hour
        m = t1.minute + t2.minute

        if m >= 60:
            h += m // 60
            m = m % 60

        return time(h,m)

t1 = time(5, 36)
t2 = time(4, 33)

ans = time.addition(t1, t2)
print(ans.hour, ":",ans.minute)
```
