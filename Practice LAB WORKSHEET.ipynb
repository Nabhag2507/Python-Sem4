## Practical 1: Student Marks Analyzer

### Problem Definition
Store student data as a list of tuples. Remove duplicate marks, calculate average marks and store result in a dictionary.

### Sample Input
```python
students = [
    (1, "Amit", [78, 85, 90, 85]),
    (2, "Neha", [88, 92, 79, 88]),
    (3, "Ravi", [70, 75, 80, 75])
]
```

### Sample Output
```
{1: ('Amit', 84.33), 2: ('Neha', 86.33), 3: ('Ravi', 75.0)}

```

````python
n = int(input("Enter total number of students: "))
stu = []
rno = 1
for i in range(n):
    templ = []
    roll = rno
    name = input("Enter student name")
    t = list(map(int, input("Enter marks :").split()))
    
    templ.append(roll)
    templ.append(name)
    templ.append(t)
    
    tup = tuple(templ)
    rno += 1
    
    stu.append(tup)

d = {}
count = 1
for i in stu:
    name = i[1]
    marks = list(set(i[2]))
    avg = sum(marks) / len(marks)
    
    li = []
    li.append(name)
    li.append(avg)
    
    d[count] = tuple(li)
    count += 1
d
````

## Practical 2: Inventory Management

### Problem Definition
Remove out-of-stock items and calculate total inventory value.

### Sample Input
```python
inventory = {
    'pen': (10, 50),
    'book': (100, 0),
    'pencil': (5, 100)
}
```

### Sample Output
```
{'pen': (10, 50), 'pencil': (5, 100)}
Total Value: 1000
```

````python
inventory = {}

n = int(input("Enter total number of items :"))
for i in range(n):
    name = input("Enter item name")
    li = []
    price = int(input("Enter price : "))
    count = int(input("Enter count : "))
    
    li.append(price)
    li.append(count)
    
    inventory[name] = tuple(li)
    
newd = {}
for i in inventory:
    if(inventory[i][1] != 0):
        newd[i] = inventory[i]
newd
````

## Practical 3: Email Domain Extractor

### Problem Definition
Extract email domains and count occurrences.

### Sample Input
```python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']
```

### Sample Output
```
{'gmail.com': 2, 'yahoo.com': 1}
```

````python
emails = []

n = int(input("Enter total no of emails :"))
for i in range(n):
    t = input("Enter email :")
    emails.append(t)
    
ans = {}

for email in emails:
    domain = email.split('@')[1]
    
    ans[domain] = ans.get(domain, 0) + 1

print(ans)
````

## Practical 4: Shopping Cart Billing

### Problem Definition
Calculate item-wise cost and total bill.

### Sample Input
```
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}
```

### Sample Output
```
{'apple': 100, 'banana': 30}
Total Bill: 130
```

````python
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}

ansd = {}
totalBill = 0

for i in cart:
    item = i[0]
    price = prices[item]
    totalBill += (i[1] * price)
    
    ansd[item] = i[1] * price

print(ansd)
print(totalBill)
````

## Practical 5: Character Position Index

### Problem Definition
Store positions of each character in a string.

### Sample Input
```python
s = "banana"
```

### Sample Output
```
{'b': [0], 'a': [1, 3, 5], 'n': [2, 4]}
```

````python
def get_character_positions(s):
    pos_dict = {}
    
    for index, char in enumerate(s):
        if char in pos_dict:
            pos_dict[char].append(index)
        else:
            pos_dict[char] = [index]
            
    return pos_dict

s = input("Enter any string :")
result = get_character_positions(s)
print(result)
````

## Practical 6: Survey Data Aggregation

### Problem Definition
Count responses city-wise and choice-wise.

### Sample Input
```python
responses = [
    ('Amit','Rajkot','Yes'),
    ('Neha','Rajkot','No'),
    ('Ravi','Ahmedabad','Yes')
]
```

### Sample Output
```
{'Rajkot': {'Yes': 1, 'No': 1}, 'Ahmedabad': {'Yes': 1}}
```

````python
data = [('Amit','Rajkot','Yes'), ('Neha','Rajkot','No'), ('Ravi','Ahmedabad','Yes')]

output = {}
for name, city, status in data:
    if city not in output:
        output[city] = {}
    
    output[city][status] = output[city].get(status, 0) + 1

print(output)
````
