### 01) Write a function to calculate BMI given mass and height. (BMI = mass/h**2)

```python
def calculateBmi(mass, height):
    return (mass / height ** 2)

mass = float(input("Enter mass :"))
height = float(input("Enter height :"))

BMI = calculateBmi(mass, height)
print(BMI)
```

### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
def primeDetection(n) -> bool:
    for i in range(int(n/2),1,-1):
        if n % i == 0:
            return 0
    return 1

n = int(input("Enter any number :"))
primeDetection(n)
```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
def primeBetweenRange(num1, num2):
    a = min(num1, num2)
    b = max(num1, num2)
    li = []
    for i in range(a, b+1):
        for j in range(int(i/2),1,-1):
            if i % j == 0:
                break
        else:
            li.append(i)
    return li

n1 = int(input("Enter first number :"))
n2 = int(input("Enter second number :"))

primeBetweenRange(n1, n2)
```

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python
def nthTermFibonacci(n):
    if n <= 3:
        return 1
    return nthTermFibonacci(n-1) + nthTermFibonacci(n-2)

nthTermFibonacci(5)
```

### 10) Write a function to get the sum of the scores ending with zero.

### 11) Write a function to invert a given Dictionary.
#### hint: keys to values & values to keys
#### Before : {'a': 10, 'b':20, 'c':30, 'd':40}
#### After : {10:'a', 20:'b', 30:'c', 40:'d'}

```python
def reverseDictionary(d):
    return {value: key for key, value in d.items()}

reverseDictionary(d)
```

### 12) Write a function that returns the number of uppercase and lowercase letters in the given string. 
#### example : Input : s1 = AbcDEfgh ,Ouptput : no_upper = 3, no_lower = 5

```python
def upperLower(s):
    ucounter = lcounter = 0
    for i in s:
        if str.isupper(i): ucounter+=1
        else : lcounter+=1
    return ucounter, lcounter

s = input("Enter any string :")
upperLower(s)
```

### 14) For the given list of names of students, extract the names having more that 7 characters. Use filter().

```python
def studentNames(s):
    if len(s) >= 7:
        return True
    return False

li = list(map(str, input("Enter student names :").split()))
list(filter(studentNames, li))
```
