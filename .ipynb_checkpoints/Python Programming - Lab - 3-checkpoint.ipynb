### 05) WAP to check whether the given year is leap year or not.
#### [If a year can be divisible by 4 but not divisible by 100 then it is leap year but if it is divisible by 400 then it is leap year]

```python
year = int(input("Enter any year :"))

print("Year is leap year") if ((year % 100 != 0 and year % 4 == 0 ) or year % 400 == 0) else print("Year is not leap year")
```

### 07) WAP to implement simple calculator which performs (add,sub,mul,div) of two no. based on user input.

```python
a = int(input("Enter first number :"))
b = int(input("Enter second number :"))
c = input("Enter your choice :")

def calc(a,b,c) -> int:
    match c :
        case '+':
            return a + b
        case '-':
            return a - b
        case '*':
            return a * b
        case '/':
            return a / b

print(f"{a} {c} {b} = {calc(a,b,c)}")
```

### 08) WAP to read marks of five subjects. Calculate percentage and print class accordingly. 
#### Fail below 35 
#### Pass Class between 35 to 45 
#### Second Class
#### between 45 to 60
#### First Class between 60 to 70
#### Distinction if more than 70

```python
a = int(input("Enter marks of first subject :"))
b = int(input("Enter marks of second subject :"))
c = int(input("Enter marks of third subject :"))
d = int(input("Enter marks of fourth subject :"))
e = int(input("Enter marks of fifth subject :"))

perc = (a + b + c + d + e) / 5

if perc > 70:
    print("Distinction")
elif perc > 60:
    print("First class")
elif perc > 45:
    print("Second class")
elif perc > 35:
    print("Pass class")
else:
    print("Fail class")
```

### 10) WAP to calculate electricity bill based on following criteria. Which takes the unit from the user.
#### a. First 1 to 50 units – Rs. 2.60/unit
#### b. Next 50 to 100 units – Rs. 3.25/unit
#### c. Next 100 to 200 units – Rs. 5.26/unit
#### d. above 200 units – Rs. 8.45/unit

```python
unit = int(input("Enter units :"))

if unit > 200:
    bill = unit * 8.45
elif unit > 100:
    bill = unit * 5.26
elif unit > 50:
    bill = unit * 3.25
else:
    bill = unit * 2.60

print(f"Your bill for {unit} units is {bill}")
```
